### Графики тренда, сезонности и шумов всего временного ряда

Построим графики тренда, сезонности и шумов за весь промежуток времени. Для этого воспользуемся функцией `seasonal_decompose()` из модуля `tsa.seasonal` библиотеки `statsmodels`.

```python
def decomposed_plot(df):
    decomposed = seasonal_decompose(df['num_orders'])

    decomposed.trend.plot(figsize=(20,12), grid=True)
    plt.title('Trend')
    plt.show()

    decomposed.seasonal.plot(figsize=(20,12), grid=True)
    plt.title('Seasonality')
    plt.show()

    decomposed.resid.plot(figsize=(20,12), grid=True)
    plt.title('Residuals')
    plt.show()
```

```python
decomposed_plot(data_analysis[:])
```

По графику тренда отчетливо прослеживается тенденция увеличения числа заказов такси. Очень сложно оценивать сезоность по всему графику, поэтому выделим небольшой фрагмент времени.

## Обучение

### Создание признаков

Добавим следующие признаки в наш DataFrame:
1. Календарные признаки (месяц, день недели, день, час);
2. «Отстающие значения»;
3. Скользящее среднее.

Так как при обучении модели мы можем менять количество признаков отстающих значений и менять ширину окна при вычислении скользящего среднего, то напишем функцию для быстро изменения DataFrame.

```python
data_ml = data.copy()
```

```python
def make_features(data, max_lag, rolling_mean_size):
    #data['month'] = data.index.month
    #data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
       data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

В первом приближении выберем max_lag = 4, rolling_mean_size = 12.

```python
make_features(data_ml, 4, 12)
data_ml.head()
```

Удалим пропущенные значения.

```python
data_ml = data_ml.dropna()
data_ml.head()
```

### Разделение выборок

Разделим данные на две выборки:
* Тренировочная - 90%;
* Тестовая - 10%.
   

Функция `train_test_split()` из модуля `sklearn.model_selection` по умолчанию перемешивает данные. Поэтому укажем аргумент shuffle равным False, чтобы разделить данные корректно. И проверим, что разделение прошло правильно.

```python
train, test = train_test_split(data_ml, shuffle=False, test_size=0.1)
print(train.index.min(), train.index.max())
print(test.index.min(), test.index.max())
```

```python
X_train = train.drop('num_orders', axis = 1)
X_test  = test.drop('num_orders', axis = 1)

y_train = train['num_orders'];
y_test  = test['num_orders'];
```

### Обучение моделей

Обучим 4 модели:
* RandomForestRegressor;
* DecisionTreeRegressor;
* CatBoostRegressor;
* LGBMRegressor.

Подбор гиперпараметров будем осуществлять с помощью `GridSearchCV()`, в качестве сплитера будем использовать `TimeSeriesSplit()`, так как он рандомно не перемешивает данные, что важно для обучения моделей временных рядов.

```python
#tscv = TimeSeriesSplit(n_splits=5)
tscv = TimeSeriesSplit(n_splits=8)
```

```python
data_result_list = []
```

#### Скорер

Напишем функцию для вычисления RMSE и создадим скорер.

```python
def rmse(y_true, y_pred):
    return mean_squared_error(y_true, y_pred) ** 0.5
scoring = make_scorer(rmse, greater_is_better=False)
```

#### RandomForestRegressor

```python
%%time
# Время выполнения около 8 мин
parametrs = { 'n_estimators': range (50, 151, 10),
              'max_depth': range (3,20, 2)}
model_RFR = RandomForestRegressor(random_state=12345)
grid_RFR = GridSearchCV(model_RFR, parametrs, cv=tscv, scoring=scoring)
grid_RFR.fit(X_train, y_train)
```

```python
data_result_list.append({'Модель': 'RandomForestRegressor', 'RMSE': abs(grid_RFR.best_score_), 'Гиперпараметры': grid_RFR.best_params_})
data_result_pd = pd.DataFrame(data_result_list)
data_result_pd
```

#### DecisionTreeRegressor

```python
%%time
# Время выполнения около 3сек
parametrs = { 'max_depth': range (3,50, 2)}
model_DTR = DecisionTreeRegressor(random_state=12345)
grid_DTR = GridSearchCV(model_DTR, parametrs, cv=tscv, scoring=scoring)
grid_DTR.fit(X_train, y_train)
```

```python
data_result_list.append({'Модель': 'DecisionTreeRegressor', 'RMSE': abs(grid_DTR.best_score_), 'Гиперпараметры': grid_DTR.best_params_})
data_result_pd = pd.DataFrame(data_result_list)
data_result_pd
```

#### CatBoostRegressor

```python
%%time
# Время выполнения около 19мин
parametrs = { 'iterations': range(100, 201, 10), 'depth': range(3, 11, 2)}
model_CBR = CatBoostRegressor(random_state=12345, verbose=False)
grid_CBR = GridSearchCV(model_CBR, parametrs, cv=tscv, scoring=scoring)
grid_CBR.fit(X_train, y_train)
```

```python
data_result_list.append({'Модель': 'CatBoostRegressor', 'RMSE': abs(grid_CBR.best_score_), 'Гиперпараметры': grid_CBR.best_params_})
data_result_pd = pd.DataFrame(data_result_list)
data_result_pd
```

#### LGBMRegressor

```python
%%time
# Время выполнения около 2 мин
parametrs = { 'n_estimators': range(10, 91, 20), 'max_depth': range(3, 9, 2), 'learning_rate': [0.01, 0.1, 1.0]}
model_LGBM = LGBMRegressor(random_state=12345)
grid_LGBM = GridSearchCV(model_LGBM, parametrs, cv=tscv, scoring=scoring)
grid_LGBM.fit(X_train, y_train)
```

```python
data_result_list.append({'Модель': 'LGBMRegressor', 'RMSE': abs(grid_LGBM.best_score_), 'Гиперпараметры': grid_LGBM.best_params_})
data_result_pd = pd.DataFrame(data_result_list)
data_result_pd
```

В результате обучения моделей и подбора гиперпараметров на тренировочных данных лучшей моделью оказалась **LGBMRegressor** с значением **RMSE = 24.78** с гиперпараметрами: **'learning_rate': 0.1, 'max_depth': 7, 'n_estimators': 100**.

## Тестирование

### Проверка моделей на тестовых данных

```python
rmse_model_test = []
```

```python
for model, best_param in [(RandomForestRegressor, grid_RFR), (DecisionTreeRegressor, grid_DTR),
                         (CatBoostRegressor, grid_CBR), (LGBMRegressor, grid_LGBM)]:
    if model == CatBoostRegressor:
        best_model = model(**best_param.best_params_, random_state=12345, verbose=False)
    else:
        best_model = model(**best_param.best_params_, random_state=12345)
    best_model.fit(X_train, y_train)
    predict = best_model.predict(X_test)
    rmse_test = rmse(y_test, predict)
    rmse_model_test.append(rmse_test)
```

```python
data_result_pd['RMSE_test'] = pd.Series(rmse_model_test, index=data_result_pd.index)
data_result_pd
```

Значения метрики RMSE на тестовых данных для всех моделей упали, но лучшей все равно остается модель LGBMRegressor с значением RMSE = 45.44.

По заданию требовалось, что значение RMSE должно быть меньше 48. Выбранная модель обеспечивает такое значение метрики RMSE.

```python
plt.figure(figsize=(20,12))
plt.plot(y_test.index, y_test, 'r-', 
         y_test.index, predict, 'b-')
plt.title('target vs predict', fontsize=12)
plt.legend(['target','predict'], fontsize=12)
plt.ylabel('Количество заказов такси', fontsize=12)
plt.xlabel('Время', fontsize=12)
plt.grid(True)
plt.xlim('2018-08-15','2018-08-19')
plt.show()
```

По графику видно, что модель плохо предсказывает пиковые значения (> 200 заказов).

### Проверка модели на адекватность

Проверим значение метрики RMSE для стандартной модели.  
В качестве предсказаний возьмем среднее значение и значение медианы по всем предсказаниям из тестовой и валидационной выборкам.

```python
predict_adequacy_mean = pd.Series(y_train.mean(), index=y_test.index)
print('Значение метрики RMSE для стандартной модели (среднее значение): {:.2f}'.format(rmse(y_test, predict_adequacy_mean)))
predict_adequacy_median = pd.Series(y_train.median(), index=y_test.index)
print('Значение метрики RMSE для стандартной модели (медиана): {:.2f}'.format(rmse(y_test, predict_adequacy_median)))
```

В результате видим, что выбрананя модель лучше чем стандартные модели (средее значение и медиана), соответственно, наша модель адекватна.
