### Расчет показателя эффективности для операций флотации и очистки и сравнение с представленными значениями в DataFrame

#### Показатель эффективности для операции флотации

Рассчитаем показатель эффективности (Recovery) для операции флотации и сравним это значение с представленным в DataFrame с помощью метрики MAE.

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']

data_train['rougher.output.recovery_calc'] = ((C * (F - T)) / (F * (C - T))) * 100
```

```python
data_train[['rougher.output.recovery_calc', 'rougher.output.recovery']].head()
```

```python
print('MAE для операции флотации: {:.3f}'
      .format(mean_absolute_error(data_train[data_train['rougher.output.recovery'].notnull()]['rougher.output.recovery'], 
                    data_train[data_train['rougher.output.recovery'].notnull()]['rougher.output.recovery_calc'])))
```

Значение MAE = 0, значит представленное значение эффективности для операции флотации рассчитано верно.

#### Показатель эффективности для операции очистки

Рассчитаем показатель эффективности (Recovery) для операции очистки и сравним это значение с представленным в DataFrame с помощью метрики MAE.

```python
C = data_train['final.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['final.output.tail_au']

data_train['final.output.recovery_calc'] = ((C * (F - T)) / (F * (C - T))) * 100
```

```python
data_train[['final.output.recovery_calc', 'final.output.recovery']].head()
```

```python
print('MAE для операции очистки: {:.3f}'
      .format(mean_absolute_error(data_train[data_train['final.output.recovery'].notnull()]['final.output.recovery'], 
                    data_train[data_train['final.output.recovery'].notnull()]['final.output.recovery_calc'])))
```

Значение MAE = 0, значит представленное значение эффективности для операции очистки рассчитано верно.

### Концентрация металлов (Au, Ag, Pb) на различных этапах очистки

Напишем функцию для построения графика концентрация металлов (Au, Ag, Pb) на различных этапах очистки.

```python
def plot_concentrate(element):
    columns = ['rougher.input.feed_'+element.lower(), 'rougher.output.concentrate_'+element.lower(), 
               'primary_cleaner.output.concentrate_'+element.lower(), 'final.output.concentrate_'+element.lower()] 
    for column in columns:
        data_full[column].hist(alpha = .5, bins = 100, edgecolor = 'white', figsize = (20,12))
    plt.legend(columns, fontsize=16)
    plt.title('Концентрация ' + element + ' на различных этапах очистки', fontsize=16)
    plt.xlabel('Концентрация', fontsize=16)
    plt.ylabel('Количество', fontsize=16)
    plt.show()
    display(data_full[columns].describe())
```

#### Концентрация золота (Au)

```python
plot_concentrate('Au')
```

## Модель

### Функция для расчета sMAPE

Напишем функцию для расчета симметричнго среднего абсолютного процентного отклонения (sMAPE). И функцию для расчета итогового знаечния sMAPE.

```python
def smape(target, predict):
    return (((abs(target - predict) / ((abs(target) + abs(predict)) / 2)).sum()) / target.shape[0]) * 100 
def total_smape (smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

### Стандартизация данных

Уберем целевые признаки из **data_train** и **data_test**.

```python
X_train = data_train.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis = 1)
X_test = data_test.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis = 1)
```

Приведем признаки к единому масштабу, для этого произведем стандартизацию данных с помощью `StandardScaler()`. Модель обучим только на тестовой выборке.

```python
numeric = X_train.columns
scaler = StandardScaler()
scaler.fit(X_train[numeric])

X_train[numeric] = scaler.transform(X_train[numeric])
X_test[numeric] = scaler.transform(X_test[numeric])
```

Предлагаю сделать две модели, одну для предсказания ***rougher.output.recovery***, другую для предсказания ***final.output.recovery***. Модель для предсказания ***rougher.output.recovery*** будем строить только на признаках, относящиеся к ***rougher***, так как все остальные признаки не имеют отношения к флотации, они выполняются после нее. Модель для предсказания ***final.output.recovery*** будем строить, используя все признаки.

```python
X_train_rougher = X_train[[name for name in X_train.columns if name.startswith('rougher')]]
y_train_rougher = data_train['rougher.output.recovery']
y_train_final = data_train['final.output.recovery']

X_test_rougher = X_test[[name for name in X_test.columns if name.startswith('rougher')]]
y_test_rougher = data_test['rougher.output.recovery']
y_test_final = data_test['final.output.recovery']
```

### Построение моделей

Для нашей задачи будем использовать 3 модели:
* DecisionTreeRegressor;
* RandomForestRegressor;
* LinearRegression.

```python
models = [DecisionTreeRegressor(random_state=12345), 
          RandomForestRegressor(random_state=12345, n_estimators=10), 
          LinearRegression()]
```

Создадим свой скорер, осованный на функции smape, которую мы написали ранее.

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

Посчитаем значения **sMAPE_rougher**, **sMAPE_final**, **sMAPE_total** для трех моделей.

```python
df_result = pd.DataFrame()
for model in models:
    cross_val_score_rougher = cross_val_score(model, 
                                              X_train_rougher, 
                                              y_train_rougher, 
                                              cv=5, 
                                              scoring=smape_scorer).mean()
    cross_val_score_final = cross_val_score(model, 
                                              X_train, 
                                              y_train_final, 
                                              cv=5, 
                                              scoring=smape_scorer).mean()
    
    data = {'sMAPE_rougher' : cross_val_score_rougher, 
           'sMAPE_final' : cross_val_score_final,
           'sMAPE_total' : total_smape(cross_val_score_rougher, cross_val_score_final)}
    df_result = df_result.append(data, ignore_index=True)

df_result.set_axis(['Decision Tree Regressor', 
                    'Random Forest Regressor', 
                    'Linear Regression'],
                    inplace=True)
df_result
```

Из результатов видно, что для **rougher** лучшее значение **sMAPE_rougher = 7.66** показала модель **Linear Regression**, а для **final** лучшее значение **sMAPE_final = 11.19** показала модель **Random Forest Regressor**.

### Подбор гиперпараметров для лучшей модели.

Попробуем улучшить значение метрики для **final**, подбирая гиперпараметры. Для этого воспользуемся функцией `GridSearchCV`.

```python
%%time
# Время выполнения около 4 минут
parametrs = { 'n_estimators': range (10, 21, 5),
              'max_depth': range (3,10, 2)}
model = RandomForestRegressor(random_state=12345)
grid_final = GridSearchCV(model, parametrs, cv=5, scoring=smape_scorer)
grid_final.fit(X_train, y_train_final)
print(grid_final.best_params_)
print(grid_final.best_score_)
```

Лучшее значение метрики **sMAPE_final = 9.60** показала модель со следующими гипермараметрами:

```python
total_smape_test = total_smape(df_result.loc['Linear Regression'].min(), grid_final.best_score_)
print('Значение метрики total_sMAPE для тренировочной выборки: {:.2f}'.format(total_smape_test))
```

### Проверка модели на тестовой выборке и проверка модели на адекватность.

Проверим качество модели на тестовой выборке.

Проверим значение метрики total_sMAPE для стандартной модели. В качестве предсказаний возьмем среднее значение по всем предсказаниям из тестовой выборки.
