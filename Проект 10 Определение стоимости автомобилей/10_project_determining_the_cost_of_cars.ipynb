### Обработка пропущенных значений

Проверим в каких признаках есть пропущенные значения, узнаем их количество и долю.

```python
def missing_values_table(df, print_result):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mz_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mz_table = mz_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mz_table['Data Type'] = df.dtypes
        mz_table = mz_table[
            mz_table.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        if print_result:
            print ("Dataframe has " + str(df.shape[1]) + " columns and " + str(df.shape[0]) + " Rows.\n"      
            "There are " + str(mz_table.shape[0]) +
              " columns that have missing values.")
        return mz_table

missing_values_table(data_clean, 1)
```

## Обучение моделей

### Кодирование Ordinal Encoding

Так как количество моделей машин достаточно большое, то лучше применять Ordinal Encoding, чем OHE, чтобы не плодить большое количесвто признаков и тем самым уменьшить время обуения модели.

```python
categorials = list(data_clean.select_dtypes(include=['object']).columns)

encoder = OrdinalEncoder() 
encoder.fit(data_clean[categorials]) 

data_clean[categorials] = encoder.transform(data_clean[categorials])
data_clean.head(10)
```

### Разделение выборок

Разделим исходную выборку на обучающую и тестовую выборки в следующей пропорции:
* Обучающая выборка - 60%;
* Валидационная выборка - 20%;
* Тестовая выборка - 20%.

```python
df_train, df_valid_test = train_test_split(data_clean, test_size=0.4, random_state=12345)
df_valid, df_test = train_test_split(df_valid_test, test_size=0.5, random_state=12345)
```

Проверим размерности выборок.

```python
print('Размерность df_train:', len(df_train))
print('Размерность df_valid:', len(df_valid))
print('Размерность df_test:', len(df_test))
```

Извлечем признаки и целевой признак для каждой выборки.

```python
X_train = df_train.drop('price', axis = 1)
X_valid = df_valid.drop('price', axis = 1)
X_test  = df_test.drop('price', axis = 1)

y_train = df_train['price']
y_valid = df_valid['price']
y_test  = df_test['price']
```

### Стандартизация данных

Приведем признаки к единому масштабу, для этого произведем стандартизацию данных с помощью `StandardScaler()`. Модель обучим только на тестовой выборке, чтобы уменьшить переобучение.

```python
numeric = ['registration_year', 'power', 'kilometer']
scaler = StandardScaler()
scaler.fit(X_train[numeric])

X_train[numeric] = scaler.transform(X_train[numeric])
X_valid[numeric] = scaler.transform(X_valid[numeric])
X_test[numeric] = scaler.transform(X_test[numeric])
```

### Построение моделей

```python
def rmse(clf, X, y):
    y_pred = clf.predict(X)
    return mean_squared_error(y, y_pred) ** 0.5
```

#### LinearRegression

```python
model_LR = LinearRegression()
start = time.time()
model_LR.fit(X_train, y_train)
time_LR_train = time.time() - start
```

```python
start = time.time() 
rmse_LR_valid = rmse(model_LR, X_valid, y_valid)
time_LR_valid = time.time() - start
```

```python
data_result_list = []
data_result_list.append({'Модель': 'LinearRegression', 'Время обучения, с': time_LR_train,
                   'Время предсказания, c': time_LR_valid, 'RMSE_valid': rmse_LR_valid, 'Гиперпараметры': '-'})
data_result_pd = pd.DataFrame(data_result_list)
data_result_pd
```

#### CatBoostRegressor

```python
%%time
# Время выполнения около 4 мин
for iteration in range(10, 111, 20):  
    for depth in range(3, 11, 2):
    
        model_CBR = CatBoostRegressor(random_state=12345, iterations=iteration, depth=depth, verbose=False)
    
        start = time.time()
        model_CBR.fit(X_train, y_train)
        time_CBR_train = time.time() - start

        start = time.time() 
        rmse_CBR_valid = rmse(model_CBR, X_valid, y_valid)
        time_CBR_valid = time.time() - start
        
#         print(depth, iteration, time_CBR_train, time_CBR_valid, rmse_CBR_valid)
        data_result_list.append({'Модель': 'CatBoostRegressor', 'Время обучения, с': time_CBR_train,
                   'Время предсказания, c': time_CBR_valid, 'RMSE_valid': rmse_CBR_valid, 
                                 'Гиперпараметры': {'iteration' : iteration, 'depth' : depth}})
data_result_pd = pd.DataFrame(data_result_list)
data_result_pd        
```

#### LightGBM

```python
%%time
# Время выполнения около 2 мин 30 сек
for n_estimators in range(10, 91, 20):  
    for max_depth in range(3, 9, 2):
        for rate in [0.01, 0.1, 1.0]:
    
            model_LGBM = LGBMRegressor(random_state=12345, n_estimators=n_estimators, max_depth=max_depth, learning_rate=rate)

            start = time.time()
            model_LGBM.fit(X_train, y_train)
            time_LGBM_train = time.time() - start

            start = time.time() 
            rmse_LGBM_valid = rmse(model_LGBM, X_valid, y_valid)
            time_LGBM_valid = time.time() - start

#             print(rate, max_depth, n_estimators, time_LGBM_train, time_LGBM_valid, rmse_LGBM_valid)
            data_result_list.append({'Модель': 'LGBMRegressor', 'Время обучения, с': time_LGBM_train,
                       'Время предсказания, c': time_LGBM_valid, 'RMSE_valid': rmse_LGBM_valid, 
                                     'Гиперпараметры': {'rate' : rate, 'n_estimators' : n_estimators, 'depth' : max_depth}})
data_result_pd = pd.DataFrame(data_result_list)
```
